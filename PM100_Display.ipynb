```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.dates as mdates
import matplotlib as mpl
import re

mpl.rcParams['figure.dpi'] = 150
folder = r'C:\Users\sk88\Downloads\606_power_fluctuations'
```

```python
pow_file_0530 = '20230530_995_seed_ON'
temp_file_0530 = '0530_temp'

pow_data_0530 = pd.read_csv(folder + '\\' + pow_file_0530 + '.csv', header=12, delimiter=';')
temp_data_0530 = pd.read_csv(folder + '\\' + temp_file_0530 + '.csv', header=0, encoding = "ISO-8859-1", delimiter=',')

temp_dates_0530 = temp_data_0530['Time (hh:mm:ss)']

temp_times_0530 = [x.split()[1] for x in temp_dates_0530]
temperature_0530 = temp_data_0530['Temperature']
temp_list_0530 = [re.findall('\d+\.*\d*', value) for value in temperature_0530]
temperature_0530 = [float(x) for sublist in temp_list_0530 for x in sublist]

power_comma_to_decimal_a_0530 = [x.replace(',','.') for x in pow_data_0530['Power (W)']]
power_float_a_0530 = [float(x) for x in power_comma_to_decimal_a_0530]
power_a_0530 = [x*1e3 for x in power_float_a_0530]

power_comma_to_decimal_b_0530 = [x.replace(',','.') for x in pow_data_0530['Power (W).1']]
power_float_b_0530 = [float(x) for x in power_comma_to_decimal_b_0530]
power_b_0530 = [x*1e3 for x in power_float_b_0530]

pow_times_0530 = list(pow_data_0530['Time of day (hh:mm:ss) '])

fig_1, ax_1 = plt.subplots()
ax_12 = ax_1.twinx()
ax_1.plot(pow_times_0530,power_a_0530,'b',label='995 towards PPLN')
ax_1.set_ylabel('995 Power Towards PPLN (mW)',color='b')
ax_1.set_xlabel('Date (hh:mm:ss)')

ax_12.plot(pow_times_0530,power_b_0530,'r',label='995 towards OSA')
ax_12.set_ylabel('995 Power Towards OSA (mW)',color='r')

ax_1.xaxis.set_major_locator(mdates.DayLocator(interval=2000))
ax_1.xaxis.set_tick_params(labelrotation=45)

fig_1.autofmt_xdate()

plt.title('995 Polarization Change')
plt.xlabel('Date')

#plt.savefig(folder + '\\' + pow_file, dpi = 150,bbox_inches='tight')
```

```python
fig_2 = plt.figure()
ax_21 = fig_2.add_subplot(111, label="1")
ax_22 = fig_2.add_subplot(111, label="2", frame_on=False)

ax_21.plot(pow_times_0530 ,power_a_0530,'b',label='995 towards PPLN')
ax_21.set_ylabel('995 Power Towards PPLN (mW)',color='b')
ax_21.set_xlabel('Date (hh:mm:ss)')
ax_21.xaxis.set_major_locator(mdates.DayLocator(interval=2000))
ax_21.xaxis.set_tick_params(labelrotation=45)

ax_22.plot(temp_times_0530, temperature_0530, 'r', label='Temperature')
ax_22.set_ylabel('Temperature $^{\circ}$ C',color='r')
ax_22.set_xlabel('Date (hh:mm:ss)')
ax_22.xaxis.tick_top()
ax_22.yaxis.tick_right()
ax_22.xaxis.set_label_position('top') 
ax_22.yaxis.set_label_position('right') 
ax_22.xaxis.set_major_locator(mdates.DayLocator(interval=50))
ax_22.xaxis.set_tick_params(labelrotation=45)
```
