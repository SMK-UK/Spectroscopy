```python
import pandas as pd
import matplotlib.pyplot as mp
```

```python
file_path = r"C:\Users\sk88\Desktop\995_AOM.csv"
excel_file = pd.read_csv(file_path)
```

```python
Pin = excel_file['Pin'][0]
freq = excel_file['Frequency']
single = excel_file['single']
double = excel_file['double']
```

```python
mp.figure()
mp.plot(freq, single, 'x')
mp.plot(freq, double, 'o')

mp.figure()
mp.plot(freq, single/Pin, 'x')
mp.plot(freq, double/Pin, 'x')
```
